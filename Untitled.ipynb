```python
from pylab import plot, show, grid, xlabel, ylabel, xlim, ylim
from numpy import exp, sin, cos, arange
import numpy as np
```

```python
g = 9.8 # m/s^2
m = 68.1 # kg
v = 40. # m/s

def fc(t, c):
    return (g*m)/c * (1-exp(- (c/m) * t)) - v
```

```python
ft = []
t = 10. # s
for i in range(1,30):
    ft.append(fc(t, i))
```

```python
plot(ft, 'o')
xlabel("c")
ylabel("f(c)")
grid()
show()
```
