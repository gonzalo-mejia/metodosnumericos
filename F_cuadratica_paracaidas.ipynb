graficar la solucion analitica y la numerica del jercicio del paracaidista

```python
from pylab import plot, show, legend
from math import exp
```

```python
vnum=[]
tp=[]
vana=[]

dlt=1
g=9.8
m=68.1
c=12.5
vi=0

for t in range (21):
    
    vi=vi+((c/m)*vi-g)*dlt
    v=(m*g/c)*(-exp(c*t/m)+1)
    
    vnum.append (vi)
    tp.append (t)
    vana.append (v)
    


plot (tp,vnum)
plot (tp,vana)
show
```
